```python
#split data 

x = df.drop('Class' , axis = 1)
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train , y_test = train_test_split(x,y, test_size = 0.2 , random_state = 42 )
```

```python
y_train.shape
```

```python
#Logestic regression (before under sampling ) 

from sklearn.linear_model import LogisticRegression 
```

```python
lg = LogisticRegression()   # (before under sampling ) 
lg.fit(x_train,y_train)
```

```python
#lets predict the value  (before under sampling ) 

y_pred1= lg.predict(x_test)
```

```python
#accuracy sore  (before under sampling ) 
accuracy_score(y_test,y_pred1)
```

```python
#some additional socre  (before under sampling ) 

print('precision_score = ',precision_score(y_test,y_pred1))
print('f1_score =',f1_score(y_test,y_pred1))
print('recall_score = ',recall_score(y_test,y_pred1))
```
